# [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)[1]

[1]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    s_x2 = 0
    s_xy = 0
    s_x = 0
    s_y = 0
    
    for i in range(0,len(X)):

        s_x += X[i]
        s_y += Y[i]

        x_2 = X[i]**2
        xy = X[i]*Y[i]
        s_x2 += x_2
        s_xy += xy
            
    m = ((len(X)*s_xy) - (s_x*s_y))/((len(X)*s_x2 - (s_x**2)))
    b = (s_y - (m*s_x))/len(X)
    
    # To Do
    
    return m, b

# # solution
m, b = best_fit(X, Y)
print("y = ",b, "+",m,"x")

#best fit line:
#y = 0.80 + 0.92x

# # To Do: plot points and fit line
plt.scatter(X,Y)
x_float = np.array(X) + 0. 
yfit = b + m*x_float
print(yfit)

plt.plot(X,yfit)
````

# [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

````python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    # To Do - Calculate m, h

    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))

````
